```python
from PIL import Image
import glob
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def change_size(directory):
    file_dir = directory.split('/')[0]
    length = len(file_dir)
    for filename in glob.glob(directory):
        _img = filename.split('.')
        img = Image.open(filename).convert('RGB').resize((105,105))
        img_name = _img[0][length+1:]
        img_format = _img[1]
        img.save('{}.{}'.format(img_name,img_format), img.format)
        img.close()
```

```python
change_size('wellsfargo/*')
```

```python
# 为了改变图片格式
i = 0
for filename in glob.glob('boa/*'):
    _img = filename.split('.')
    img = Image.open(filename).convert('RGB').resize((105,105))
    img_name = '.'.join(_img)
    print(img_name)
    img.save('{}.{}'.format(i, _img[1]), img.format)
    img.close()
    i+= 1
```
